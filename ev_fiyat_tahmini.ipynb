```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dosya = '/content/drive/MyDrive/train.csv'
df= pd.read_csv(dosya)
df.head(3)
```

```python
#eksik verileri gormek icin
eksikler =df.isnull().sum()
eksikler[eksikler>0]

sayisal_sutunlar = df[eksikler.index].select_dtypes(include=['float64','int64']).columns
kategorik_sutunlar = df[eksikler.index].select_dtypes(include=['object']).columns

for sutun in sayisal_sutunlar:
  df[sutun] = df[sutun].fillna(df[sutun].median())

for sutun in kategorik_sutunlar:
    df[sutun] = df[sutun].fillna(df[sutun].mode()[0])
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df, columns=df.select_dtypes(include=["object"]).columns, dtype=int)
```

```python
df.select_dtypes(include=["object"]).columns
```

```python
#x sorular ozellikler ,y cevap hedefler

y = df["SalePrice"]
x = df.drop("SalePrice", axis=1)

```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

print(x_train.shape)
print(x_test.shape)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

```

```python
#tahminler
y_pred = model.predict(x_test)
y_pred[:5]
```

```python
#karsılastırma
print("tahmin", y_pred[:5])
print("gercek", y_test[:5])
```

```python
#mea(mean absolute errror): tahminler ortalamadan kaç dolar saptı
#r^2 model veriyi ne kadar iyi açıklıyor

mae = mean_absolute_error(y_test.values, y_pred)
r2 = r2_score(y_test.values, y_pred)

print("MAE:", mae)
print("R²:", r2)
```

```python
df["TotalSF"] = df["TotalBsmtSF"] + df["1stFlrSF"] + df["2ndFlrSF"]
df["HouseAge"] = 2026 - df["YearBuilt"]
df[["TotalSF","HouseAge"]].head()
```

```python
#yeniden x ve y olusturmalıyım
y = df["SalePrice"]
x = df.drop("SalePrice",axis=1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
model = LinearRegression()
model.fit(x_train,y_train)
#tahmin yap
y_pred = model.predict(x_test)
#degerlendir
mae = mean_absolute_error(y_test.values, y_pred)
r2 = r2_score(y_test.values, y_pred)
print("MAE:", mae)
print("R²:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_rf = RandomForestRegressor()
model_rf.fit(x_train, y_train)
y_pred_rf = model_rf.predict(x_test)

mae_rf = mean_absolute_error(y_test.values, y_pred_rf)
r2_rf = r2_score(y_test.values, y_pred_rf)

print("MAE:", mae_rf)
print("R²:", r2_rf)
```
